4th dataset (house price dataset)

```python
from sklearn.datasets import load_boston
house_data=load_boston()
house_feature=pd.DataFrame(house_data.data,columns=house_data.feature_names)
house_target=house_data.target
```

```python
house_feature.head()
```

```python
house_feature["PRICE"]=house_target
```

<h1>Displot</h1>

```python
sns.displot(house_feature["PRICE"])
```

<h1>Correlation Heat Map</h1>
<h4>Positive correlation / Negative Correlation<h4>

```python
correlation=house_feature.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt="1f",annot=True,annot_kws={"size":6},cmap="Greens")
```
